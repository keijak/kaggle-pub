```python
# TODO: Parse these two columns.
# - list feature
# - dict keys are enclosed with single quotes.
# - dict keys are sometimes enclosed with double quotes (when the string contains single quote) 

bad_json_columns = ['customDimensions', 'hits']
json_columns = ['device', 'geoNetwork', 'totals', 'trafficSource']

def parse_json(df):
    print('total rows: ', len(df))
    df = df.assign(rowId=pd.Series(df.index))
    for col in json_columns:
        print('processing', col)
        def f(row: pd.Series):
            if row.rowId % 10000 == 0:
                print(f'progress: {100 * row.rowId // len(df)}%')
            s = row[col]
            try:
                json_data = json.loads(s)
            except json.JSONDecodeError as e:
                print('json error at row {}'.format(row.rowId))
                print(s)
                raise e
            keys, vals = ['rowId'], [row.rowId]
            if isinstance(json_data, list):
                keys.append(col + '_count')
                vals.append(len(json_data))
            else:
                for key, val in json_data.items():
                    if not isinstance(val, dict):
                        keys.append('_'.join([col, key]))
                        vals.append(val)
                        continue
                    for key2, val2 in val.items():
                        k = '_'.join([col, key, key2])
                        if isinstance(val2, dict):
                            # Only trafficSource_adwordsClickInfo_targetingCriteria is deeply nested
                            # Apparently, it's always empty.
                            if val2:
                                print('too deeply nested structure:', k)
                            continue
                        keys.append(k)
                        vals.append(val2)
            return pd.Series(vals, index=keys)

        parsed = df.apply(f, axis=1)
        df = df.drop(columns=[col], axis=1, errors='ignore')
        df = df.merge(parsed, how='left', on='rowId')

    return df
```

```python
df_train_small = df_train_full.head(100)
df_train_small = parse_json(df_train_small)
df_train_small.head().T
```

```python
df_train_small.drop(columns=bad_json_columns, axis=1, inplace=True)
df_train_small.head().T
```

```python
# Strip & parse test data.
df_test_stripped = df_test_full.drop(columns=bad_json_columns, axis=1)
df_test_parsed = parse_json(df_test_stripped)
df_test_parsed.head().T
```
